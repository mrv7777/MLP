Write a code to predict the house price of California Housing dataset using GridSearchCV.

Write your code based on the following keypoints:

Split the California housing dataset into train and test set with 70:30 ratio with

 random_state = 1.

Import StandardScaler for scaling X_train and X_test to X_train_norm and X_test_norm

 with_mean = True
 and
 with_std = True

Import SGDRegressor with

 random_state = 1
Pass SGDRegressor through GridSearchCV
Hyperparamter tuning to be done over

loss
 as 'squared_error' or 'huber'

penalty
 as 'l1' or 'l2'

alpha
 as 0.1, 0.01, 0.001
maximum number of passes as [1000,2000,5000]
Cross Validation = 4
Train the 'model' and compute the 'score' on test data

```python
featurs, lable = fetch_california_housing(return_X_y=True, as_frame=True)
```

```python
X_train,X_test,Y_Train,Y_test = train_test_split(featurs,lable,test_size=0.3,random_state=1)
```

```python
sgd_reg_pipeline = Pipeline([("feature_scaling", StandardScaler(with_mean=True,with_std=True)),
                             ("sgd_reg", SGDRegressor(random_state = 1,))])
```

```python
param_grid = {
    'sgd_reg__loss': ['squared_error', 'huber'],
    'sgd_reg__penalty': ['l1', 'l2'],
    'sgd_reg__alpha': [0.1, 0.01, 0.001],
    'sgd_reg__max_iter': [1000, 2000, 5000]
}
```

```python
grid_search = GridSearchCV(sgd_reg_pipeline, param_grid, cv=4)
```

```python
grid_search.fit(X_train, Y_Train)
```


1- Enter the value of the 'score'.(Enter your answer in four decimal places)

```python
y_pred = grid_search.predict(X_test)
```

```python
score = r2_score(Y_test,y_pred)
```

```python
grid_search.best_params_
```
